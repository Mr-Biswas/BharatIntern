```python
# Looping through the 'Published_date' column to convert string dates to datetime objects, 
# and then extracting the year, month, day, and date into separate lists for further processing.
# Importing datetime library which provides great functions to handle date and time information
import datetime
i=0
for i in range(data.shape[0]):
    date_time_obj = datetime.datetime.strptime(data['Published_date'].at[i], '%Y-%m-%d')
    data['Published_date'].at[i] = date_time_obj
    i = i+1
i=0
date=[]
year=[]
month=[]
day=[]
for i in range(data.shape[0]):
    d = data['Published_date'][i].date();
    y = data['Published_date'][i].date().year;
    m = data['Published_date'][i].date().month;
    da = data['Published_date'][i].date().day;
    date.append(d) # Storing dates
    year.append(y) # Storing years
    month.append(m) # Storing months
    day.append(da) # Storing days
    i = i+1
data.drop(['Published_date'], inplace=True,axis=1)
data['Published_date']=date
data['year']=year
data['month'] = month
data['day'] = day
```

# Exploratory Analysis and Visualization
In this section,will see different relationships among channel statistics parameters ( views, likes, comments, dislikes etc).

```python
#Total views, likes, and comments of all videos
data[['Views','Likes','Comments']].sum()
```

```python
# Get the average (mean) of 'Likes', 'Views', and 'Comments' columns from the 'describe()' summary statistics
AvgLikes = data.describe()['Likes']['mean']
AvgViews = data.describe()['Views']['mean']
AvgComments = data.describe()['Comments']['mean']
print('Average number of views on video = ',AvgViews,'\nAverage number of likes on video = ',AvgLikes,'\nAverage number of comments on video = ',AvgComments,'\n')
```

```python

# Create a new figure for the pie charts
fig = plt.figure()

# First pie chart: Viewers who react on videos
ax1 = fig.add_axes([0, 0, 0.75, 0.75], aspect=1)  
# Add axes to the figure at specified position [left, bottom, width, height], with aspect ratio 1:1
# Pie chart showing viewers who reacted to the video (Liked) vs neutral viewers (No reaction)
pie_vars = ['Reacters', 'Neutral']  
# Labels for the pie chart
pie_values = [data['Likes'].sum(), data['Views'].sum() - (data['Likes'].sum())]  
# Values: Total likes and the remaining views (neutral viewers)
ax1.pie(pie_values, labels=pie_vars, autopct='%1.2f%%')  
# Plot the pie chart and display percentage values
ax1.set_title('Viewers who react on video')  
# Set the title of the chart

# Second pie chart: Type of reacters (Likers vs Commenters)
ax2 = fig.add_axes([0.8, 0, 0.75, 0.75], aspect=1)  
# Add another axes for the second pie chart
# Pie chart showing the proportion of viewers who liked the video vs those who commented
pie_vars = ['Likers', 'Commenters']  
# Labels for the pie chart
pie_values = [data['Likes'].sum(), data['Comments'].sum()]  
# Values: Total likes and total comments
ax2.pie(pie_values, labels=pie_vars, autopct='%1.2f%%')  
# Plot the pie chart with percentages
ax2.set_title('Type of reacters')  
# Set the title of the chart

# Third pie chart: Commenters vs Non-commenters with respect to total viewers
ax3 = fig.add_axes([0.4, -0.75, 0.75, 0.75], aspect=1)  
# Add another axes for the third pie chart at a different position
# Pie chart showing commenters vs non-commenters based on total viewers
pie_vars = ['Comments', 'Non-Commenters']  
# Labels for the pie chart
pie_values = [data['Comments'].sum(), data['Views'].sum() - data['Comments'].sum()] 
# Values: Total comments and non-comments (Viewers who didn't comment)
ax3.pie(pie_values, labels=pie_vars, autopct='%1.2f%%')  
# Plot the pie chart with percentages
ax3.set_title('Viewers vs total comments')  
# Set the title of the chart

# Display the figure with all three pie charts
plt.show() 
# Show the plots
```

```python
data['Published_date'] = pd.to_datetime(data['Published_date'], format='%y-%m-%d')

# Step 2: Convert 'Published_date' to integer in 'YYMMDD' format using integer extraction
data['Published_date_int'] = (data['Published_date'].dt.year % 100).astype(str) + \
                              data['Published_date'].dt.month.astype(str).str.zfill(2) + \
                              data['Published_date'].dt.day.astype(str).str.zfill(2)

# Convert to integer type
data['Published_date_int'] = data['Published_date_int'].astype(int)

# Print the result
print(data)
```
