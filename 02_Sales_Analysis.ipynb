# EDA on GENDER COLUMN

```python
# Created a bar plot showing the count of each category in the 'Gender' column of the DataFrame 'df'.
a=sn.countplot(x = 'Gender', data= df, color='green')
```

```python
## Created a count plot of the 'Gender' column in the dataframe 'df' with orange bars and label each bar with its corresponding count.
a=sn.countplot(x = 'Gender', data= df, color='orange')

for bars in a.containers:
    a.bar_label(bars)
```

```python
#Group the dataframe by 'Gender', sum the 'Amount' for each group, and sort the results in descending order of 'Amount'.
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# Group the dataframe by 'Gender', sum the 'Amount' for each group, sort the results in descending order of 'Amount', 
# and then create a bar plot to visualize the total 'Amount' for each 'Gender'.
sales_gen= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.barplot(x= 'Gender',y= 'Amount' ,data=sales_gen)
```

# EDA On AGE Group

```python
df.columns
```

```python
a= sn.countplot(data=df, x= 'Age Group', hue= 'Gender')

for bars in a.containers:
    a.bar_label(bars)
```

```python
#Grouped the dataframe by 'Age Group', sum the 'Amount' for each group, sort the results in descending order of 'Amount', 
# and then create a bar plot to visualize the total 'Amount' for each 'Age Group'.

sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.barplot(x= 'Age Group',y= 'Amount' ,data=sales_age, color= "green")
```

# EDA on STATE Column

```python
# Group the dataframe by 'State', sum the 'Orders' for each state, sort the results in descending order of 'Orders', 
# take the top 10 states, and create a bar plot to visualize the total 'Orders' for these top 10 states with a custom figure size.

sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sn.set(rc={'figure.figsize':(15,5)})
sn.barplot(data= sales_state, x= "State", y= "Orders", color="yellow")
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sn.set(rc={'figure.figsize':(15,5)})
sn.barplot(data= sales_state, x= "State", y= "Amount")
```

# EDA on OCCUPATION

```python
#Created a count plot for the 'Occupation' column, set the figure size to (10,5), and add labels showing the count for each bar.
# Set the figure size using Matplotlib's plt.figure()
plt.figure(figsize=(20, 5))
a= sn.countplot(data=df, x= 'Occupation', width=0.5, color="purple")

sn.set(rc={'figure.figsize':(15,8)})
for bars in a.containers:
    a.bar_label(bars, fontsize=11)
```

```python
sales_state=df.groupby(['Occupation',], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sn.set(rc={'figure.figsize':(20,5)})
sn.barplot(data= sales_state, x= "Occupation", y= "Amount", color="brown")
```
